```python
import pandas as pd 
import os
```

```python
folder = 'data'
files = os.listdir(folder)
files
```

```python
departments_frame = pd.read_csv(os.path.join(folder,files[0]))
departments_frame
```

```python
dept_emp_frame = pd.read_csv(os.path.join(folder,files[1]))
dept_emp_frame['emp_pos'] = 'employee'
dept_emp_frame
# Merge on dept_no
```

```python
dept_manager_frame = pd.read_csv(os.path.join(folder,files[2]))
dept_manager_frame['emp_pos'] = 'manager'
dept_manager_frame
# if emp_no starts with 11, they are a manager
# Merge on dept_no 
```

```python
# Concat dept_emp_frame and dept_manager_frame to get full employee frame
dept_employees = pd.concat([dept_emp_frame,dept_manager_frame],axis=0)
dept_employees
```

```python
# Create master frame from three frames to merge with employees frame on emp_no
# Start with depertments_frame and dept_emp_frame
dept_frame = pd.merge(departments_frame,dept_employees,on='dept_no',how='right')
dept_frame
```

```python
employees_frame = pd.read_csv(os.path.join(folder,files[3]))
employees_frame.head()
```

```python
employee_master_frame = pd.merge(dept_frame,employees_frame,on='emp_no',how='left')
employee_master_frame
```

```python
salaries_frame = pd.read_csv(os.path.join(folder,files[4]))
salaries_frame.head()
```

```python
employee_master_frame = pd.merge(employee_master_frame,salaries_frame,on='emp_no',how='left')
employee_master_frame
```

```python
titles_frame = pd.read_csv(os.path.join(folder,files[5]))
titles_frame.head()
```

```python
employee_master_frame = pd.merge(employee_master_frame,titles_frame,left_on='emp_title_id',right_on='title_id',how='left')
employee_master_frame
```
